````python
## Importando las librerias
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
````

````python
### Leer los datos
oecd_bli = pd.read_csv("oecd_bli.csv", thousands=',')
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
oecd_bli.head(5)
````

````python
gdp_per_capita = pd.read_csv("gdp_per_capita.csv",thousands=',',delimiter='\t',
                             encoding='latin1', na_values="n/a")
gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
gdp_per_capita.set_index("Country", inplace=True)
gdp_per_capita['GDP per capita'].head(10)
````

````python
gdp_per_capita[gdp_per_capita.index == 'Guatemala']
````

#### Preparando los datos

Vamos a unir las tablas a modo de tener en una tabla el pais GDP, y el nivel de satisfaccion de vida

````python
full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita, left_index=True, right_index=True)
full_country_stats.sort_values(by="GDP per capita", inplace=True)
full_country_stats[["GDP per capita", 'Life satisfaction']].head(3)
````

##### Dividimos la data

````python
remove_indices = [0, 1, 6, 8, 33, 34, 35]
keep_indices = list(set(range(36)) - set(remove_indices))

variables = ['GDP per capita', 'Life satisfaction', 'Homicide rate', 'Air pollution', 'Employment rate']

sample_data = full_country_stats[variables].iloc[keep_indices]
missing_data = full_country_stats[variables].iloc[remove_indices]
````

````python
sample_data
````

### Graficando la data con Matplotlib

````python
sample_data.plot(kind='scatter', x='GDP per capita', y='Life satisfaction')
````

### Graficando la data con Seaborn

Seaborn agrega usa una regresegion linear con los datos de entrada, si no queremos que grafique la linea agregamos el parametro ```fit_reg=False ```

````python
sns.regplot(x=sample_data['GDP per capita'], y=sample_data['Life satisfaction'])
````

````python
sns.regplot(x=sample_data['GDP per capita'], y=sample_data['Life satisfaction'], fit_reg=False)
````
